# Pregunta 1
Dado el siguiente texto: 

texto = """ 

Juan: 987654321 

María: 912345678 

Pedro: 998877665 

Ana: 923456789 

"""

Realice lo siguiente:

a) Extraiga todos los números telefónicos utilizando expresiones regulares.

b) Muestre la cantidad total de números encontrados.



## Respuesta a) y b)

```python
#Se importa el paquete re para usarlo:
import re

#Se añade el texto a utilizar
texto = """
Juan: 987654321
María: 912345678
Pedro: 998877665
Ana: 923456789
"""
#Se usa "\b9" para asegurar que el número comienze en 9
#Se usa "\d" para escoger secuencias de dígitos o números
#Se usa "\d{8}" para que luego del iniciador "9" sigan 8 digitos correspondientes a los telefonos "+51" que comienzan con 9 y tienen 9 digitos.

#Respuesta a)
telefonos= re.findall(r"\b9\d{8}\b", texto)

#Respuesta b)
#Se imprime los números válidos
print("Números telefónicos válidos:",telefonos)
#Se imrpime el total de números encontrados
print("Cantidad total de números encontrados", len(telefonos))
```
