```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "http://books.toscrape.com/"
# Realizar la petición GET
response = requests.get(url)
# Verificar el código de estado
if response.status_code == 200:
    print("Conexión exitosa. Código de estado:", response.status_code)
else:
    print("Error en la conexión. Código de estado:", response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
#print(soup)
```

```python
libros = soup.select("article.product_pod")
#print(libros)
```

```python
# Lista para almacenar la información
libros_list = []

for libro in libros:
    # Nombre del libro
    nombre = libro.find("h3").find("a")["title"]
    #print(nombre)
    
    # Precio
    precio_sucio = libro.find("p", class_="price_color").get_text()
    
    # Forzar codificación a UTF-8
    # Transformamos el texto mal interpretado (latin-1) a bytes y lo decodificamos en utf-8
    try:
        precio = precio_sucio.encode('latin-1').decode('utf-8')
    except (UnicodeEncodeError, UnicodeDecodeError):
        # Método alternativo de limpieza por si la recodificación estricta falla
        precio = precio_sucio.replace('Â', '').strip()
    
    libros_list.append(
        {
            "nombre": nombre,
            "precio": precio,
            }
    )

print("LInformación de los primeros 20 libros mostrados en la página principal")
for item in libros_list:
    print(f"Libro: {item['nombre']} | Precio: {item['precio']}")
```

```python
import pandas as pd
import re

# Asumimos que la variable 'libros_list' ya existe y tiene los datos de la Pregunta 3.
# Ejemplo de lo que contiene:
# libros_list = [{'nombre': 'Libro A', 'precio': '£51.77'}, {'nombre': 'Libro B', 'precio': '£10.50'}]

# a) Cree un DataFrame utilizando pandas.
df_libros = pd.DataFrame(libros_list)

# b) Guarde la información en un archivo llamado: libros.csv
# index=False evita que se guarde la columna de índices numéricos de Pandas
df_libros.to_csv("libros.csv", index=False, encoding='utf-8')
```

```python
# c) Determine: El libro más caro y el libro más barato.
# Primero creamos una columna auxiliar numérica para poder comparar los precios.
# Usamos 're.sub' para eliminar cualquier caracter que NO sea un dígito (\d) o un punto (.).

df_libros['precio_numerico'] = df_libros['precio'].apply(lambda x: float(re.sub(r'[^\d.]', '', x)))

# Encontramos el índice del libro con el precio máximo y mínimo.
indice_mas_caro = df_libros['precio_numerico'].idxmax()
indice_mas_barato = df_libros['precio_numerico'].idxmin()

# Extraemos las filas completas utilizando los índices obtenidos.
libro_mas_caro = df_libros.loc[indice_mas_caro]
libro_mas_barato = df_libros.loc[indice_mas_barato]

# 4. Mostramos el resultado en pantalla (imprimiendo el nombre y su precio original)
print("RESULTADO")
print(f"El libro más caro es: '{libro_mas_caro['nombre']}' con un precio de {libro_mas_caro['precio']}")
print(f"El libro más barato es: '{libro_mas_barato['nombre']}' con un precio de {libro_mas_barato['precio']}")
```
